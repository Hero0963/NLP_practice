```python
from simpletransformers.classification import ClassificationModel, ClassificationArgs
import pandas as pd
```

```python
df = pd.read_csv("imdb_sample/texts.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns=["labels","text","is_valid"]
```

```python
df["labels"]=df["labels"].apply(lambda s: 0 if s == "negative" else 1)
```

```python
df
```

```python
df = df[["text","labels","is_valid"]]
train_df=df[~df["is_valid"]]
eval_df=df[df["is_valid"]]
```

```python
train_df.shape,eval_df.shape
```

```python
# Optional model configuration
model_args = ClassificationArgs(num_train_epochs=30, evaluate_during_training=True, overwrite_output_dir=True, 
                               fp16=False,max_seq_length=256, do_lower_case=True, multiprocessing_chunksize=1, )

# Create a ClassificationModel
model = ClassificationModel(
    'bert',
    './bert-tiny-unpretrained',
    num_labels=2,
    args=model_args,
    # use_cuda=False
) 
```

```python
model.train_model(train_df,eval_df=eval_df)
```

```python
# Evaluate the model
result, model_outputs, wrong_predictions = model.eval_model(eval_df)
```

```python
result
```

```python
from sklearn.metrics import classification_report
```

```python
pred, _ = model.predict(eval_df["text"].tolist())
```

```python
pred
```

```python
print(classification_report(eval_df["labels"],pred))
```
